### Resolução

````python
# Definindo os atributos necessários p/ Quandl
CODIGOS_SERIES = ['BCB/1253', 'BCB/433', 'BCB/11']

# Carregar os dados do Quandl
macro_br = quandl.get(dataset = CODIGOS_SERIES,
                      authtoken = API_KEY,
                      collapse = "quarterly")
# Renomear as colunas
macro_br.columns = ['PIB', 'IPCA', 'SELIC']

# Criar uma coluna para selic em taxa
macro_br['TAXA_SELIC'] = macro_br['SELIC'] * 100

# Visualizando dataframe
macro_br
````

````python
# Calculando estatística descritiva
macro_br[['IPCA', 'TAXA_SELIC']].describe()
````

````python
# Dataviz (Gráfico de linha)

# Dimensões do nosso gráfico
fig, ax = plt.subplots(figsize = (20,8))

# Construção do gráfico
ax = sns.lineplot(data = macro_br['PIB'],
                  color = 'Darkgreen')

# Definir nomes dos eixos
ax.set(xlabel = 'Ano', 
       ylabel = 'Índice')


# Definir título
ax.axes.set_title('Produto Interno Bruto do Brasil - 1996 a 2015', fontsize = 26)


# Formato da caixa do gráfico
sns.despine(offset = 10, trim = True)

# Diminuir a opacidade do grid
ax.grid(alpha = 0.2)
````

````python
# Dataviz (Gráfico de linha)

# Dimensões do nosso gráfico
fig, ax = plt.subplots(figsize = (20,8))

# Construir nosso gráfico
ax = sns.lineplot(data = macro_br[['IPCA', 'TAXA_SELIC']],
                  dashes = False,
                  markers = False)

# Definir nomes dos eixos
ax.set(xlabel = 'Ano', 
       ylabel = 'Percentual (%)')


# Definir título
ax.axes.set_title('IPCA e Selic do Brasil - 1980 a 2022', fontsize = 26)


# Formato da caixa do gráfico
sns.despine(offset = 10, trim = True)

# Diminuir a opacidade do grid
ax.grid(alpha = 0.2)
````

### Resolução OPTATIVA

````python
# Filtrar as observações apenas para depois dos anos 2000
macro_br_2000 = macro_br.loc[(macro_br.index > "31-12-1999") & (macro_br.index < '01-01-2022')]

# Selecionar ipca e selic
macro_br_2000 = macro_br_2000[['IPCA', 'TAXA_SELIC']]

# Visualiza essas informações
macro_br_2000
````

````python
# Criar uma defasagem
macro_br_2000['TAXA_SELIC_LAG'] = macro_br_2000['TAXA_SELIC'].shift(-1)

# Retirando valores ausentes
macro_br_2000 = macro_br_2000.dropna()

macro_br_2000
````

````python
import scipy.stats # Para calcular a correlação

# Calculou correlação entre ipca e selic
cor_ipca_selic = scipy.stats.pearsonr(macro_br_2000['IPCA'], macro_br_2000['TAXA_SELIC'])[0]

print(f"Correlação do IPCA e Taxa Selic: {cor_ipca_selic:.2}")
````

````python
# Calculou correlação entre ipca e selic defasada
cor_ipca_selic_lag = scipy.stats.pearsonr(macro_br_2000['IPCA'], macro_br_2000['TAXA_SELIC_LAG'])[0]

print(f"Correlação do IPCA e Lag Selic: {cor_ipca_selic_lag:.2}")
````

````python
# Criando acumulação de 4 trimestres
macro_br_2000['SELIC_LAG_ACU'] = macro_br_2000['TAXA_SELIC_LAG'].rolling(window = 4).sum()
macro_br_2000['IPCA_ACU'] = macro_br_2000['IPCA'].rolling(window = 4).sum()

# Retirando valores ausentes
macro_br_2000 = macro_br_2000.dropna()

macro_br_2000
````

````python
# Calculou correlação entre ipca acumulado e selic defasada acumulada
cor_ipca_acu_selic_lag_acu = scipy.stats.pearsonr(macro_br_2000['IPCA_ACU'], macro_br_2000['SELIC_LAG_ACU'])[0]

print(f"Correlação do IPCA acumulado e Lag Selic acumulada: {cor_ipca_acu_selic_lag_acu:.2}")
````

````python
# Dataviz (Histogramas)

# Dimensões do nosso gráfico
fig, ax = plt.subplots(figsize = (20,8))

# Criar um gráfico

ax = sns.histplot(macro_br_2000['IPCA'])


# Definir nomes dos eixos
ax.set(xlabel = 'Valores', 
       ylabel = 'Frequência')


# Definir título
ax.axes.set_title('Distribuição do IPCA', fontsize = 26)


# Formato da caixa do gráfico
sns.despine(offset = 10, trim = True)

# Diminuir a opacidade do grid
ax.grid(alpha = 0.2)
````
