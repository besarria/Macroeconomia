# Alteração da taxa de juros

```python
x = np.arange(0,2000,0.1) 
    #Usando a função arange do numpy
    #x fica no intervalo entre 0 e 4 (domínio) e vai variar de 0.5 entre os pontos
```

```python
#Plot 1: usando pylab e matplotlib

y = x
z = 0.6*x + 400
w = 0.6*x + 600
plt.plot(x,z, color='blue', linewidth=2.5, linestyle="--", label='$D_0$')
plt.plot(x,w, color='green', linewidth=2.5, linestyle="--", label='$D_1$')
plot(x,y, 'r--', label='Oferta')
#plt.ylabel('Demanda (Z), Produção (Y)')
plt.xlabel('Renda (Y)')
plt.annotate(r'$i_1 (i_1>i_0)$',
            xy=(1000, 1000), xycoords='data',
            xytext=(-60, -90), textcoords='offset points', fontsize=10,
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
plt.annotate(r'$i_0$',
            xy=(1500, 1500), xycoords='data',
            xytext=(-60, -90), textcoords='offset points', fontsize=10,
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
plt.axhline(ls='--', c='k') 
# Inclui a linha do eixo 'x'
legend(loc='best')
plt.savefig('efeito_juros')
```

```python
x = np.arange(1,5000,0.1) 
    #Usando a função arange do numpy
    #x fica no intervalo entre 0 e 4 (domínio) e vai variar de 0.5 entre os pontos
```

# Derivação da curva IS

```python
import numpy as np  # importando numpy
import matplotlib.pyplot as plt# importanto o matplotlib para a geração dos gráficos
```

```python
y = x
z = 0.6*x + 300
w = 0.6*x + 900
s = np.exp(-x/1500)

plt.figure(figsize=(6, 4))
plt.subplot(2, 1, 1)
plt.plot(x,z, color='blue', linewidth=2.5, linestyle="--", label='$Z(i_1)$')
plt.plot(x,w, color='green', linewidth=2.5, linestyle="--", label='$Z(i_0)$')
plot(x,y, 'r--')
#plt.ylabel('Demanda (Z), Produção (Y)')
plt.xlabel('Renda (Y)')
legend(loc='best')
# Inclui a linha do eixo 'x'
#plt.xticks(())
#plt.yticks(())
#plt.text(0.5, 0.5, 'subplot(2,1,1)', ha='center', va='center', size=24, alpha=.5)

plt.subplot(2, 1, 2)
plt.plot(x,s, color='green', linewidth=2.5, linestyle="--", label='Curva IS')
plt.ylabel('Taxa de juros (i)')
plt.xlabel('Renda (Y)')
plt.axhline(ls='--', c='k') 

# Inclui a linha do eixo 'x'
#plt.xticks(())
#plt.yticks(())
legend(loc='best')
plt.tight_layout()
plt.savefig('curva_IS')
plt.show()
```

# Curva LM

```python
x = np.arange(0,4,0.5)
v = np.arange(0,4,0.5)
w = np.arange(0,4,0.5)
    #Usando a função arange do numpy
    #x fica no intervalo entre 0 e 4 (domínio) e vai variar de 0.5 entre os pontos
```

```python
#Plot 3: usando matplotlib

z = 2 + 0*x
y = np.exp(-x)+1
d = np.exp(-w)+2
e = np.exp(w/15)-0.9

#plt.figure(figsize=(6, 4))
plt.subplot(1, 2, 1)

plt.plot(x,y,'rx--', label='$Y_0$')   # Demanda 01
plt.plot(w,d,'green', label='$Y_1$')  # Demanda 02
plot(z,v, 'blue')      # Oferta
#plt.title('Soluções da Eq_1')

plt.ylabel('Taxa de juros (i)')
plt.xlabel('Moeda real (M)')
#plt.legend(['Eq_1'])
plt.axhline(ls='--', c='k')
legend(loc='best')

plt.subplot(1, 2, 2)
plt.plot(w,e, color='green', linewidth=2.5, linestyle="--", label='Curva LM')
#plt.ylabel('Taxa de juros (i)')
plt.xlabel('Renda (Y)')
plt.axhline(ls='--', c='k') 

# Inclui a linha do eixo 'x'
#plt.xticks(())
#plt.yticks(())
legend(loc='best')
plt.savefig('curva_LM')
plt.show()
```

# Combinação das curvas IS e LM

```python
x = np.arange(1,6,0.1) 
    #Usando a função arange do numpy
    #x fica no intervalo entre 0 e 4 (domínio) e vai variar de 0.5 entre os pontos
```

```python
#Plot 1: usando pylab e matplotlib

a = 40+2*x
b = 50-x
plt.plot(x,a, color='blue', linewidth=2.5, linestyle="--", label='$LM$')
plt.plot(x,b, color='green', linewidth=2.5, linestyle="--", label='$IS$')
#plot(x,y, 'r--', label='Oferta')
plt.ylabel('Taxa de juros (i)')
plt.xlabel('Renda (Y)')
#plt.axhline(ls='--', c='k') 
# Inclui a linha do eixo 'x'
legend(loc='best')
plt.savefig('IS_LM')
```

```python
#Plot 1: usando pylab e matplotlib

a = 40+2*x
b = 50-x
c = 48 - x
plt.plot(x,a, color='blue', linewidth=2.5, linestyle="--", label='$LM$')
plt.plot(x,b, color='green', linewidth=2.5, linestyle="--", label='$IS (T_0)$')
plot(x,c, 'r--', label='$IS (T_1>T_0)$')
plt.ylabel('Taxa de juros (i)')
plt.xlabel('Renda (Y)')
#plt.axhline(ls='--', c='k') 
# Inclui a linha do eixo 'x'
legend(loc='best')
plt.savefig('IS_LM_T')
```
