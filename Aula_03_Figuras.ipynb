```python
x = np.arange(0,4,0.5) 
    #Usando a função arange do numpy
    #x fica no intervalo entre 0 e 4 (domínio) e vai variar de 0.5 entre os pontos
```

```python
#Plot 1: usando pylab e matplotlib

y = 1/x 
plot(x,y, 'xg:', label='Demanda')
plt.axhline(ls='--', c='k') 
    #Inclui a linha do eixo 'x'
legend(loc='best')
plt.savefig('demanda_lab')
    #Escolhe a melhor localização pra legenda
```

```python
#Plot 3: usando matplotlib

plt.plot(x,y,'rx--')
#plt.title('Soluções da Eq_1')
plt.ylabel('Preço')
# Taxa de juros (i)
plt.xlabel('Demanda')
# Moeda (M)
#plt.legend(['Eq_1'])
plt.annotate('Inclinação', xy=(2, 4.6), xytext=(1, 1.5),
            arrowprops=dict(facecolor='black', shrink=0.05),
            )
plt.axhline(ls='--', c='k')
plt.savefig('moeda_lab')
plt.show()
```

# Demanda por moeda

```python
x = np.arange(1,15,0.1) 
v = np.arange(4,15,0.1) 
    #Usando a função arange do numpy
    #x fica no intervalo entre 0 e 4 (domínio) e vai variar de 0.5 entre os pontos
```

```python
#Plot 1: usando pylab e matplotlib
y = 4/x
z = 16/v 
plt.plot(v,z, color='blue', linewidth=2.5, linestyle="--", label='$M^d_1$')
plot(x,y, 'r--', label='$M^d_0$')
#plt.ylabel('Demanda (Z), Produção (Y)')
plt.ylabel('Taxa de juros (i)')
plt.xlabel('Demanda por moeda (M)')

plt.annotate(r'$Y_1>Y_0$',
            xy=(6.4, 2.5), xycoords='data',
            xytext=(20, 70), textcoords='offset points', fontsize=10,
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

plt.annotate("",
            xy=(1.6, 2.5), xycoords='data',
            xytext=(6.4, 2.5), textcoords='data',
            arrowprops=dict(arrowstyle="<-"))

plt.axhline(ls='--', c='k') 
# Inclui a linha do eixo 'x'
legend(loc='best')
plt.savefig('demanda_moeda')
```

# Equilíbrio no mercado monetário

```python
x = np.arange(0,4,0.5)
v = np.arange(0,4,0.5)
w = np.arange(0,4,0.5)
    #Usando a função arange do numpy
    #x fica no intervalo entre 0 e 4 (domínio) e vai variar de 0.5 entre os pontos
```

```python
#Plot 1: usando pylab e matplotlib

y = 1/x 
plot(x,y, 'xg:', label='Eq_1')
plt.axhline(ls='--', c='k') 
    #Inclui a linha do eixo 'x'
legend(loc='best')
    #Escolhe a melhor localização pra legenda
```

```python
#Plot 3: usando matplotlib

z = 2 + 0*x

plt.plot(x,y,'rx--', label='$M^d$')
plot(z,v, 'blue', label='$M^s$')
#plt.title('Soluções da Eq_1')
plt.ylabel('Taxa de juros (i)')
plt.xlabel('Moeda (M)')
#plt.legend(['Eq_1'])
plt.axhline(ls='--', c='k')
legend(loc='best')
plt.savefig('equilibrio_moeda')
plt.show()
```

# Deslocamento da demanda

```python
#Plot 3: usando matplotlib

z = 2 + 0*x
d = 2.5/w

plt.plot(x,y,'rx--', label='$M^d$')   # Demanda 01
plt.plot(w,d,'green', label='$M^d$')  # Demanda 02
plot(z,v, 'blue', label='$M^s$')      # Oferta
#plt.title('Soluções da Eq_1')

plt.annotate(r'$Y_1>Y_0$',
            xy=(2, 1.25), xycoords='data',
            xytext=(20, 70), textcoords='offset points', fontsize=10,
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

plt.annotate("",
            xy=(0.8, 1.25), xycoords='data',
            xytext=(2, 1.25), textcoords='data',
            arrowprops=dict(arrowstyle="<-"))


plt.ylabel('Taxa de juros (i)')
plt.xlabel('Moeda (M)')
#plt.legend(['Eq_1'])
plt.axhline(ls='--', c='k')
legend(loc='best')
plt.savefig('equilibrio_moeda_renda')
plt.show()
```
