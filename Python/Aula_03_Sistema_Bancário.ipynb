# Sistema Bancário

## Demanda por moeda

#### A demanda por moeda é dividida em demanda por depósitos à vista e demanda por papel moeda:

#### $D^m=zM^d$
#### $D^d=(1-z)M^d$
##### sendo $D^m$ a demanda por papel-moeda; $z$ é a proporção fixa de papel-moeda; $D^d$ a demanda por depósitos; $M^d$ é a demanda por moeda $M^d$ é a demanda por moeda.

## Demanda por reservas

#### Admitindo que as reservas bancárias são representadas por:

#### $R^d=\theta(1-z)M^d$
##### sendo $\theta$ a proporção fixa que os bancos deverão manter na forma de reservas.

## Demanda por moeda do Banco Central

#### A demanda por moeda do Banco Central $(H^d)$ é definida por:

#### $H^d=D^m+R^d$

#### substituindo as demandas por papel-moeda e reservas, temos:

#### $H^d=[z+\theta(1-z)]M^d$

#### Igualando a oferta de moeda $(H^s)$ a demanda por moeda, temos:

#### $H^s=H^d$

#### $H^s=[z+\theta(1-z)]M^d$

# Simulando o multiplicador de moeda

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

D     = 100   # Depósito 
z = 0.9       # Taxa de reserva bancária é 0.10
n_rounds = 50 # Número de períodos que vamos considerar

# Cria uma matriz de números de 0 a 50, um para cada rodada de depósitos
t = np.arange(0,n_rounds)

# Resolve a equação abaixo para cada rodada individual
S = D*(z**t)
```

# Gráfico

```python
plt.bar(t,S, color='r',label='Depósitos')
plt.grid()
plt.legend(loc='center right')
plt.xlabel('Períodos, t')
plt.ylabel('Depósitos')

plt.tight_layout()
plt.show()
```

# Variação na oferta de moeda

```python
# Solução da equação para cada período individual
M = D*(1-(z)**(t+1))/(1-z)
```

# Gráfico

```python
plt.bar(t,M, label='Variação na oferta de moeda')
plt.grid()
plt.legend(loc='center right')
plt.xlabel('Períodos, t')
plt.ylabel('Variação oferta moeda')

plt.tight_layout()
plt.show()
```

# Gráfico conjunto

```python
plt.bar(t,M, label='Variação na oferta de moeda')
plt.bar(t,S, color='r',label='Depósitos')
plt.grid()
plt.legend(loc='center right')
plt.xlabel('Períodos, t')
plt.ylabel('Oferta de moeda')

plt.tight_layout()
plt.show()
```
